```python
#Listing columns of interest
interest_columns = ['Term', 'LoanStatus', 'BorrowerAPR', 'BorrowerRate','ProsperRating (Alpha)','Occupation','ListingCategory (numeric)', 'BorrowerState','EmploymentStatus','IsBorrowerHomeowner','IncomeRange','StatedMonthlyIncome','LoanOriginalAmount','LoanOriginationDate', 'Investors']
```

```python
loan_data = df_copy[interest_columns]
```

```python
loan_data.info()
```

```python
#removing Null values from ProsperRating and Occupation.
loan_data = loan_data.dropna(subset=['ProsperRating (Alpha)'])
loan_data = loan_data.dropna(subset=['Occupation'])

#Convert LoanOriginationDate to datetime datatype.
loan_data['LoanOriginationDate'] = pd.to_datetime(loan_data['LoanOriginationDate'])
```

```python
loan_data.info()
```

## Univariate Exploration

> 

```python
#Plotting loan amount Distribution
bin = np.arange(5000, loan_data.LoanOriginalAmount.max()+300, 300)
plt.figure(figsize=[10, 5])
plt.hist(data = loan_data,  x='LoanOriginalAmount', bins = bin);
plt.xlabel('Original loan amount ($)');
plt.title('Loan Original amount Distribution');
```

>We can observe from the large spikes that most of the loans are multiples of 5K.

```python
#plotting loan Status Distribution
plt.figure(figsize = [10, 5]) 
base_color = sb.color_palette()[0]
LoanStatus_order = loan_data['LoanStatus'].value_counts().index
LoanStatus_counts = loan_data['LoanStatus'].value_counts()
n_LoanStatus = loan_data['LoanStatus'].value_counts().sum()
sb.countplot(data=loan_data, y='LoanStatus', color=base_color, order=LoanStatus_order);
plt.ylabel('Loan Status')
plt.title('Loan-Status Distribution');

# Logic to print the proportion text on the bars
for i in range (LoanStatus_counts.shape[0]):
    count = LoanStatus_counts[i]
    # Convert count into a percentage, and then into string
    pct_string = '{:0.1f}%'.format(100*count/n_LoanStatus)
    # Print the string value on the bar. 
    plt.text(count+1, i, pct_string, va='center')
```

>The majority are `current loans `and the second big part are `completed loans`.

```python
loan_data.IncomeRange.value_counts()
```

```python
# Convert IncomeRange ordered categorical types
Income_Range= ['$100,000+', '$75,000-99,999', '$50,000-74,999', '$25,000-49,999', '$1-24,999', '$0']
ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = Income_Range)
loan_data['IncomeRange'] = loan_data['IncomeRange'].astype(ordered_var)
```

```python
#plotting Borrower Incomes ranges Distribution
plt.figure(figsize = [10, 5]) 
sb.countplot(data=loan_data, y='IncomeRange', color=base_color, order=Income_Range);
plt.ylabel('Income Range')
plt.title(' Borrower Incomes ranges Distribution');
```

>Number of Borrowers whose income between `$25K-49K` and between `$50K-74K` represent the largest part followed by `$100K+`

```python
#plotting Borrower Monthly Incomes Distribution.
plt.figure(figsize = [10, 5])
bins = np.arange(0, 50000, 500)
plt.hist(data=loan_data, x='StatedMonthlyIncome', bins=bins);
plt.xlabel('Monthly income ($)')
plt.title('Borrower Monthly Income Distribution');
```

>Borrower monthly income right screwed with peek around `5K`

```python
#plotting Employment Status Distribution.
plt.figure(figsize = [10, 5]) 
base_color = sb.color_palette()[0]
EmploymentStatus_order = loan_data['EmploymentStatus'].value_counts().index
EmploymentStatus_counts = loan_data['EmploymentStatus'].value_counts()
n_EmploymentStatus = loan_data['EmploymentStatus'].value_counts().sum()
sb.countplot(data=loan_data, y='EmploymentStatus', color=base_color, order=EmploymentStatus_order);
plt.ylabel('Employment Status')
plt.title('Employment-Status Distribution');

# Logic to print the proportion text on the bars
for i in range (EmploymentStatus_counts.shape[0]):
    count = EmploymentStatus_counts[i]
    # Convert count into percentage then a string
    pct_string = '{:0.1f}%'.format(100*count/n_EmploymentStatus)
    # Print the string value on the bar. 
    plt.text(count+1, i, pct_string, va='center')
```

>Most borrowers are ` Employed `.

```python
#plotting Loan Terms Distribution.
sb.countplot(data=loan_data, x='Term', color=base_color);
plt.xlabel('Terms (Month)')
plt.title('Loans Terms');
```

>Seems that most of the borrowers prefer the moderated term which is `3-Years`.

```python
# Convert ProsperRating ordered categorical types
Prosper_order = ['HR','E','D','C','B','A','AA']
ordered_var = pd.api.types.CategoricalDtype(ordered = True,categories = Prosper_order)
loan_data['ProsperRating (Alpha)'] = loan_data['ProsperRating (Alpha)'].astype(ordered_var)

#plotting Prosper Rating Distribution.
sb.countplot(data=loan_data, x='ProsperRating (Alpha)', color=base_color,order=Prosper_order);
plt.xlabel('ProsperRating')
plt.title('Prosper Rating Distribution');
```

>Most of the borrowers rating is between `"D to A"` with a high `C` rating.

```python
loan_data['BorrowerAPR'].describe()
```

```python
#Adjusting Borrower APR bins on X-axis.
bins = np.arange(0, loan_data.BorrowerAPR.max()+0.05, 0.01)

#plotting Borrower APR Distribution
plt.hist(data = loan_data, x = 'BorrowerAPR', bins = bins);
plt.xlabel('Borrower APR');
plt.title('Borrower APR Distribution');
```

>The Borrowers Annual Percentage Rate (APR) for the loans varies between `4.5%` - `42.3%` with an average `21.9%`

## Bivariate Exploration


```python
plt.figure(figsize=[10,5]);
sb.barplot(data = loan_data, x = 'IncomeRange', y = 'LoanOriginalAmount', color = base_color);
plt.title('Income-Range vs Loan-Amount');
```

>As we can see, there is a positive correlation between income ranges and Loan amount as the income range increased the loan amount also increased except for the amount $0, as it was assumed before that it might be missing data

```python
plt.figure(figsize=[10,5])
sb.countplot(data = loan_data, x = 'IncomeRange', hue = 'IsBorrowerHomeowner', palette = 'Blues');
plt.xticks(rotation=15);
plt.title('Income-Range vs Home owning');
```

>As we can see, for the largest three income ranges, the  borrower with a homeowner has the highest share

```python
plt.figure(figsize=[10,5])
sb.violinplot(data = loan_data, x = 'EmploymentStatus', y='LoanOriginalAmount')
plt.xticks(rotation=15)
plt.title('Loan-Amount vs Employment-Status');
```

>The highest borrowers Employment Status from the loan amount are the `Employed`

```python
plt.figure(figsize=[20,15]);

plt.subplot(2, 2, 1)
sb.countplot(data=loan_data, x='ProsperRating (Alpha)', hue = 'EmploymentStatus', palette = 'Blues');
plt.xlabel('ProsperRating')
plt.title(' Prosper-Rating vs Employment-Status');

plt.subplot(2, 2, 2)
sb.countplot(data=loan_data, x='ProsperRating (Alpha)', hue = 'EmploymentStatus', palette = 'Blues');
plt.xlabel('ProsperRating')
plt.ylim(0,2000,250)
plt.title(' Prosper-Rating vs Employment-Status');
```

The Lowest Prosper rating are `Not-employed`, `retired`, and `part-time`

```python
plt.figure(figsize=[20,15]);

plt.subplot(2, 2, 1)
sb.countplot(data = loan_data, x = 'EmploymentStatus', hue = 'Term', palette = 'Blues')
plt.xticks(rotation=15)
plt.title('Employment-Status vs Term');

plt.subplot(2, 2, 2)
sb.countplot(data = loan_data, x = 'EmploymentStatus', hue = 'Term', palette = 'Blues')
plt.xticks(rotation=15)
plt.ylim(0,1500,150)
plt.title('Employment-Status vs Term');
```

>It appears that the `3-Year` term is the most preferable term among the Employment Status.

```python
plt.figure(figsize=[10,5]);
sb.regplot(data = loan_data, x = 'BorrowerAPR', y = 'LoanOriginalAmount', scatter_kws={'alpha':0.005});
plt.title('BorrowerAPR vs Loan-Amount');
```

>There is a negative correlation between the loan amount and borrower APR

## Multivariate Exploration


```python
plt.figure(figsize=[10,5])
sb.barplot(data = loan_data, x = 'IncomeRange', y = 'LoanOriginalAmount', hue = 'Term');
plt.xticks(rotation = 15);
plt.title('Income-Range vs Loan-Amount and Term');
```

>The loan amount and terms increase with income range

```python
fig = plt.figure(figsize = [10,5])
ax = sb.pointplot(data = loan_data, x = 'ProsperRating (Alpha)', y = 'BorrowerAPR', hue = 'Term',palette = 'Blues', linestyles = '', markers = '^',dodge = 0.2);
plt.title('Borrower-APR vs Prosper-Rating and term');
plt.ylabel('Borrower-APR');
```

>For Prosper Rating `(HR-C)`, the Loan term decreases with the increase of Borrower APR but for starting from Prosper Rating `(B)`, the loan term increase with the increasing of Borrower APR

```python
g = sb.FacetGrid(data = loan_data, col = 'Term', height = 3);
g.map(sb.boxplot, 'IsBorrowerHomeowner', 'LoanOriginalAmount');
```

>The average loan amount for a `1-year` loan term is the same for homeowners and non-homeowner and the average of homeowners starts to be greater starting from the `3-Year` loan term.
